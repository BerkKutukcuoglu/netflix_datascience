# KAYIT TÜRÜNE GÖRE VERİ ORANI

```python
sns.countplot(data_netflix["type"])
fig = plt.gcf()
fig.set_size_inches(5,5)
plt.title("kayıt türü")
```

```python
sns.countplot(data_netflix["rating"])
#yazıları dik yapalım
sns.countplot(data_netflix["rating"]).set_xticklabels(sns.countplot(data_netflix["rating"]).get_xticklabels(),rotation=90,ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title("Rating Derecelerine Göre Sayılar")
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="rating",hue="type",data=data_netflix)
plt.title("Tür ve Reyting Oranları")
plt.show()
```

```python
this_label = ["Movie","TV Show"]
size = data_netflix["type"].value_counts()
this_color = plt.cm.Wistia(np.linspace(0,1,2))
this_explode = [0,0.1]
plt.rcParams["figure.figsize"] = (9,9)
plt.pie(size,labels=this_label,colors=this_color,explode=this_explode,shadow=True,startangle=90)
plt.title("Tür Oranları Pasta Grafiği")
plt.legend()
plt.show()
```

```python
data_netflix["rating"].value_counts().plot.pie(autopct="%1.1f%%",shadow=True,figsize=(10,8))
plt.title("Rating Derece Oranı Pasta Grafiği")
plt.show()
```
